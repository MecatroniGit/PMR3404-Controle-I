```python
%matplotlib inline 
# Python packages
import numpy as np
import control.matlab as co
import matplotlib.pyplot as plt

s = co.tf('s');
G = (0.1)/(s*(3*s+1)*(2*s+1))
print('-------------')
print('FT da planta Gp(s) = ')
print(G)
Kc = 5
KG = Kc*G
co.sisotool(KG)
plt.savefig("a) MF")
```

```python
%matplotlib qt
mag, phase, omega = co.bode(G)
sinFi = np.sin((40-15.2+8)*np.pi/180)
alpha = (1-sinFi)/(1+sinFi)
print(alpha)
g = -20*np.log10(1/np.sqrt(alpha))
print(g)
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
    #return array[idx]

print(omega[find_nearest(mag, g)])
```
